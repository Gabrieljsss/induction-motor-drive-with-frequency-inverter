```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Ea = 8.854 * 10**(-12)
hl = 10
Ew = 710 * 10**(-12)
def sensor_output(hw):
    return hw * Ew + (hl - hw) * Ea
```

```python
hw_values = np.linspace(0, hl, 100)
```

```python
C_values = sensor_output(hw_values)
```

```python
plt.plot(hw_values, C_values)
plt.xlabel('hw')
plt.ylabel('C')
plt.title('Saída do Sensor')
plt.grid(True)
plt.show()
```
